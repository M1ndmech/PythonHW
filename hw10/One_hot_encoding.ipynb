```python
import pandas as pd
```

```python
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
```

```python
dummies = pd.get_dummies(data['whoAmI'], drop_first = True)
dummies
```

```python
types_list = data["whoAmI"].tolist()
def one_hot (data, list):
    iter_dummies = []
    for i in range (len(data)):
        if data.iloc[i, 0] == list[0]:
            iter_dummies.append(1)
        else: 
            iter_dummies.append(0)
    return iter_dummies
new_dummies = pd.DataFrame({types_list[0]:one_hot(data,types_list)})
new_dummies
```
